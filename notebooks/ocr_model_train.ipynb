```python
import numpy as np
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os
```

```python
# download dataset at https://universe.roboflow.com/anpr-8jutb/ocr-mrkbm/dataset/2


DATASET_PATH = "./dataset"
DATASET_PATH_TEST = DATASET_PATH + "/test"
DATASET_PATH_TRAIN = DATASET_PATH + "/train"
DATASET_PATH_VALID = DATASET_PATH + "/valid"
print(os.path.exists(DATASET_PATH))
```

```python
YOLO_MODEL_PATH = "./model/yolo26s.pt"
MODEL_FINETUNNED_PATH = "./model/ocr_model.pt"
```

```python
model = YOLO(YOLO_MODEL_PATH)
model = model.to('cpu')
model.train(data=DATASET_PATH + '/data.yaml', epochs=50, imgsz=640, batch=16, name="ocr_model")
```

```python
model.save(MODEL_FINETUNNED_PATH)
```

```python
plt.plot(model.history["metrics/box_loss"], label="box_loss")
plt.plot(model.history["metrics/cls_loss"], label="cls_loss")
plt.plot(model.history["metrics/obj_loss"], label="obj_loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curves")
plt.legend()
plt.show()
```

```python
images = model.val().plot()
plt.subplot(1, 2, 1)
plt.imshow(images[0])
plt.title("Predicted")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(images[1])
plt.title("Ground Truth")
plt.axis("off")
plt.show()
```
